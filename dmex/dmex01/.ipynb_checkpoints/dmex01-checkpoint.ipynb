```python
import pandas as pd
```

```python
# 导入同目录下的 data.csv
data = pd.read_csv('data.csv')
```

（1）判断第一列(Id)是否有缺失值：如果有，则补全；

```python
print(f"第一列中有缺失值的索引为：\n{data[data['Id'].isna()].index}")
```

```python
# 遍历行号，因为 Id 为唯一值，从1开始，行号从0开始
# 所以行号与 Id 的关系为 Id = index + 1
for i in data[data['Id'].isna()].index:
    data['Id'].iloc[i] = i+1
```

```python
# 再次查看第一列是否还有缺失值
data[data['Id'].isna()]
```

（2）判断是否有重复记录：如果有，则删除至唯一;

```python
# 刨除 id 列，便于查找重复内容
data_drop_id = data.drop(columns='Id')
# 找出重复行
duplicated_rows = data_drop_id[data_drop_id.duplicated(keep=False)]
duplicated_Id = data_drop_id[data_drop_id.duplicated()].index
```

```python
data = data.drop(index=duplicated_Id)
```

```python
data
```

（3）计算成绩的平均值，作为新的一列加入到原数据框中

```python
data['Average'] = data[['Chinese', 'Math', 'English']].mean(axis=1)
```

（4）寻找平均分最高的记录；

```python
max_avg_index = data['Average'].idxmax()
highest_avg_record = data.loc[data['Average'] == data['Average'].loc[max_avg_index]]
```

```python
highest_avg_record
```

（5）统计每个科目及格(≥60分)的人数。

```python
pass_Chinese_count = data[data['Chinese'] >= 60].shape[0]
pass_Math_count = data[data['Math'] >= 60].shape[0]
pass_English_count = data[data['English'] >= 60].shape[0]
```
