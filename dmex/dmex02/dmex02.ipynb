```python
df = pd.read_csv('Income.csv')
```

```python
df.head()
```

```python
# 计算各个项的支持度
item_support = df.mean(axis=0)

# 打印项的支持度
print("各个项的支持度：")
print(item_support)

# 单独查看项"age=14-34"和项"sex=male"的支持度
print("\n项'age=14-34'的支持度：", item_support['age=14-34'])
print("项'sex=male'的支持度：", item_support['sex=male'])

# 查看支持度最大的前 10 个项
print("\n支持度最大的前 10 个项：")
print(item_support.nlargest(10))
```
